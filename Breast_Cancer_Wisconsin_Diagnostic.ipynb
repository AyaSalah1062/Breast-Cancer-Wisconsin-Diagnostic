#K-Means:
1. Initialization:
Choose the number of clusters=k.
Randomly initialize k centroids.

2. Assignment:
Assign each data point to the nearest centroid, forming k clusters.

3. Update Centroids:
Recalculate the centroids by taking the mean of all data points in each cluster.

4. Iteration:
Repeat the assignment and centroid update steps until convergence or a maximum number of iterations is reached.

5. Convergence Check:
Check if the centroids have stopped changing significantly.



```python
# Function to initialize centroids randomly
def initialize_centroids(X, k):
    # print("Shape of X:", X.shape)  # Add this line for debugging
    if X is None or X.shape[0] < k:
        raise ValueError("Invalid input for initialization.")

    indices = np.random.choice(X.shape[0], k, replace=False)
    return X[indices]


# Function to assign each data point to the nearest centroid
def assign_to_centroids(X, centroids):
    distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
    return np.argmin(distances, axis=1)

# Function to update centroids based on the assigned data points
def update_centroids(X, labels, k):
    new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])
    return new_centroids

# Function to perform K-Means clustering
def kmeans(X, k, max_iters=100):
    centroids = initialize_centroids(X, k)

    for _ in range(max_iters):
        labels = assign_to_centroids(X, centroids)
        new_centroids = update_centroids(X, labels, k)

        # Check for convergence
        if np.array_equal(centroids, new_centroids):
            break

        centroids = new_centroids

    return labels, centroids
```


#PCA:
PCA aims to find a lower-dimensional representation of the data by identifying the directions (eigenvectors) along which the data has the maximum variance to capture the most important information with low dim.

#Steps:
1. Standardize the Data:
Scale the features to have zero mean and unit variance.

2. Compute the Covariance Matrix:
Calculate the covariance matrix of the standardized features.
(The covariance matrix captures the relationships between different features.
Pca find data with high variance)

3. Compute Eigenvectors and Eigenvalues:
Find the eigenvectors and eigenvalues of the covariance matrix.
(The eigenvectors represent the directions of maximum variance in the data)

* By finding these eigenvectors and eigenvalues, we identify the principal components of the data.
4. Sort Eigenvalues:
Sort the eigenvalues in descending order and choose the top-k eigenvectors (principal components) corresponding to the largest eigenvalues. Here, 'k' is the desired number of dimensions.

5. Create Projection Matrix:
Form a projection matrix using the selected eigenvectors.
(The projection matrix is formed by stacking the selected eigenvectors as columns. This matrix defines the new subspace in which we will project the original data. )

6. Transform Data:
Project the original data onto the new subspace formed by the principal components.
(The final step is to project the original standardized data onto the new subspace defined by the principal components. )

```python
def pca(X,components):
    # Step 1: Standardize the Data
    #->done at the beginning

    # Step 2: Compute the Covariance Matrix
    covariance_matrix = np.cov(X, rowvar=False)

    # Step 3: Compute Eigenvectors and Eigenvalues
    eigenvalues, eigenvectors = np.linalg.eig(covariance_matrix)

    # Step 4: Sort Eigenvalues
    sorted_indices = np.argsort(eigenvalues)[::-1]

    # Step 5: Select Top Eigenvectors
    selected_indices = sorted_indices[:components]
    principal_components = eigenvectors[:, selected_indices]

    # Step 6: Project Original Data
    new_data = np.dot(X, principal_components)

    return new_data
```

The SSE is the sum of the squared distances between each data point and its assigned centroid

```python
# Function to calculate the sum of square errors
def calculate_sse(X, centroids, labels):
    sse = 0
    for i in range(len(centroids)):
        cluster_points = X[labels == i]
        sse += np.sum(np.square(cluster_points - centroids[i]))
    return sse

```

```python
def elbow_method(X, max_k):
    errors = []
    for k in range(1, max_k + 1):
        labels, centroids = kmeans(X, k)
        errors.append(np.sum(np.square(X - centroids[labels])))
    return errors
```
